# 引入Pandas对象：



在最基本的层面上，Pandas对象可以被视为NumPy结构化数组的增强版本，其中行和列是通过标签而非简单的整数索引来识别的。

正如我们将在本章中看到的，Pandas在基本数据结构之上提供了一系列有用的工具、方法和功能，但接下来的几乎所有内容都需要理解这些结构是什么。

因此，在进一步讨论之前，让我们先看看这三种基本的Pandas数据结构：`Series`、`DataFrame`和`Index`。

我们将以标准的NumPy和Pandas导入开始我们的代码会话。

````python
import numpy as np
import pandas as pd
````

## Pandas系列对象

Pandas `Series` 是一个一维的索引数据数组。

可以通过以下方式从列表或数组创建：

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0])
data
````

`Series`结合了一系列值和一个明确的索引序列，我们可以通过`values`和`index`属性来访问它们。

`values`只是一个熟悉的NumPy数组。

````python
data.values
````

`index` 是一种类似数组的对象，类型为 `pd.Index`，我们将稍后详细讨论。

````python
data.index
````

像使用NumPy数组一样，可以通过熟悉的Python方括号表示法，通过相关索引访问数据。

````python
data[1]
````

````python
data[1:3]
````

尽管如此，正如我们将看到的，Pandas `Series` 比它所模拟的一维 NumPy 数组要更加通用和灵活。

### Series作为广义NumPy数组：


从我们目前所见，`Series`对象在本质上可能与一维NumPy数组基本可互换。

其根本区别在于，NumPy数组具有一个*隐式定义*的整数索引，用于访问值，而Pandas `Series`则有一个*显式定义*的与值相关联的索引。

这种显式索引定义赋予了`Series`对象额外的功能。例如，索引不必是整数，可以由任何所需类型的值组成。

因此，如果我们愿意，可以使用字符串作为索引。

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=['a', 'b', 'c', 'd'])
data
````

访问

````python
data['b']
````

我们甚至可以使用不连续或非顺序的索引：

````python
data = pd.Series([0.25, 0.5, 0.75, 1.0],
                 index=[2, 5, 3, 7])
data
````

````python
data[5]
````

### Series 作为专门化字典

因此，您可以将 Pandas 的 `Series` 看作是 Python 字典的一种专门化形式。

字典是一种将任意键映射到一组任意值的结构，而 `Series` 是一种将类型化键映射到一组类型化值的结构。

这种类型信息非常重要：正如 NumPy 数组背后的特定类型编译代码使其在某些操作中比 Python 列表更高效一样，Pandas `Series` 的类型信息也使其在某些操作中比 Python 字典更高效。

通过直接从一个 Python 字典构造一个 `Series` 对象，可以更加清晰地理解 `Series` 与字典之间的类比。以下是根据 2020 年人口普查数据列出的五个美国人口最多的州：

````python
population_dict = {'California': 39538223, 'Texas': 29145505,
                   'Florida': 21538187, 'New York': 20201249,
                   'Pennsylvania': 13002700}
population = pd.Series(population_dict)
population
````

从这里，可以执行典型的字典式项访问：

````python
population['California']
````

与字典不同，`Series` 还支持数组风格的操作，例如切片：

````python
population['California':'Florida']
````

我们将在[数据索引与选择](03.02-Data-Indexing-and-Selection.ipynb)中讨论Pandas索引和切片的一些特性。

### 构建 Series 对象

我们已经看到了一些从头构建 Pandas `Series` 的方法。所有这些方法都可以归纳为以下形式：

```python
pd.Series(data, index=index)
```

其中，`index` 是一个可选参数，而 `data` 可以是多种实体之一。

例如，`data` 可以是列表或 NumPy 数组，在这种情况下，`index` 默认为整数序列。

````python
pd.Series([2, 4, 6])
````

或者，`data` 可以是一个标量，它会被重复以填充指定的索引：

````python
pd.Series(5, index=[100, 200, 300])
````

或者它可以是一个字典，在这种情况下，`index` 默认为字典的键：

````python
pd.Series({2:'a', 1:'b', 3:'c'})
````

在每种情况下，可以明确设置索引以控制所使用的键的顺序或子集：

````python
pd.Series({2:'a', 1:'b', 3:'c'}, index=[1, 2])
````

## Pandas 数据框对象

Pandas 中下一个基本结构是 `DataFrame`。

与前一节讨论的 `Series` 对象类似，`DataFrame` 可以被视为 NumPy 数组的一种推广，或者作为 Python 字典的一种特化。

接下来，我们将分别探讨这两种观点。

### DataFrame作为广义NumPy数组

如果`Series`是具有显式索引的一维数组的类比，那么`DataFrame`就是具有显式行和列索引的二维数组的类比。

就像你可以将二维数组视为一系列对齐的一维列，您也可以将`DataFrame`视为一系列对齐的`Series`对象。

在这里，“对齐”意味着它们共享相同的索引。

为了演示这一点，让我们首先构建一个新的`Series`，列出前面章节讨论的五个州各自的面积（以平方公里为单位）：

````python
area_dict = {'California': 423967, 'Texas': 695662, 'Florida': 170312, 
             'New York': 141297, 'Pennsylvania': 119280}
area = pd.Series(area_dict)
area
````

现在我们有了这个以及之前的 `population` Series，我们可以使用字典构建一个包含这些信息的二维对象。

````python
states = pd.DataFrame({'population': population,
                       'area': area})
states
````

像 `Series` 对象一样，`DataFrame` 也具有一个 `index` 属性，该属性用于访问索引标签。

````python
states.index
````

此外，`DataFrame` 具有一个 `columns` 属性，该属性是一个 `Index` 对象，用于保存列标签。

````python
states.columns
````

因此，`DataFrame`可以被视为二维NumPy数组的推广，其中行和列都有一个通用索引用于访问数据。


### DataFrame作为专门化字典

同样，我们也可以将`DataFrame`视为字典的一种专门化。

字典将键映射到值，而`DataFrame`则将列名映射到一系列列数据。

例如，查询`'area'`属性会返回包含我们之前看到的面积的`Series`对象。

````python
states['area']
````

请注意这里可能存在的混淆点：在二维NumPy数组中，`data[0]`将返回第一行。而对于`DataFrame`，`data['col0']`将返回第一列。

因此，将``DataFrame``视为广义字典而非广义数组可能更为合适，尽管这两种看待方式都可以提供帮助。

我们将在[数据索引与选择](03.02-Data-Indexing-and-Selection.ipynb)中探讨更灵活的``DataFrame``索引方法。

### 构建 DataFrame 对象

Pandas `DataFrame` 可以通过多种方式构建。

在这里，我们将探讨几个示例。

#### 从单个 Series 对象构建

`DataFrame` 是 `Series` 对象的集合，可以从单个 `Series` 构建一个单列的 `DataFrame`。

````python
pd.DataFrame(population, columns=['population'])
````

#### 从字典列表创建

任何字典列表都可以转换为 `DataFrame`。

我们将使用简单的列表推导式来创建一些数据。

````python
data = [{'a': i, 'b': 2 * i}
        for i in range(3)]
pd.DataFrame(data)
````

即使字典中缺少某些键，Pandas 也会用 `NaN` 值（即“不是一个数字”；请参见 [处理缺失数据](03.04-Missing-Values.ipynb)）填充它们。

````python
pd.DataFrame([{'a': 1, 'b': 2}, {'b': 3, 'c': 4}])
````

#### 从 Series 对象字典构建

正如我们之前所看到的，`DataFrame` 也可以从 `Series` 对象的字典中构建。

````python
pd.DataFrame({'population': population,
              'area': area})
````

#### 从二维NumPy数组创建

给定一个二维数据数组，我们可以使用任何指定的列名和索引名来创建一个 `DataFrame`。

如果省略，将为每个元素使用整数索引。

````python
pd.DataFrame(np.random.rand(3, 2),
             columns=['foo', 'bar'],
             index=['a', 'b', 'c'])
````

#### 从 NumPy 结构化数组开始

我们在 [结构化数据：NumPy 的结构化数组](02.09-Structured-Data-NumPy.ipynb) 中讨论了结构化数组。

Pandas 的 `DataFrame` 操作方式与结构化数组非常相似，并且可以直接从一个创建。

````python
A = np.zeros(3, dtype=[('A', 'i8'), ('B', 'f8')])
A
````

````python
pd.DataFrame(A)
````

## Pandas索引对象

正如您所看到的，`Series`和`DataFrame`对象都包含一个明确的*索引*，使您能够引用和修改数据。

这个`Index`对象本身是一个有趣的结构，可以被视为*不可变数组*或*有序集合*（从技术上讲，它是一个多重集合，因为`Index`对象可能包含重复值）。

这些观点在可用于`Index`对象的操作方面具有一些有趣的影响。

作为一个简单示例，让我们从整数列表构造一个 `Index`:

````python
ind = pd.Index([2, 3, 5, 7, 11])
ind
````

### Index as Immutable Array

`Index` 在许多方面的操作方式类似于数组。

例如，我们可以使用标准的 Python 索引表示法来检索值或切片。

````python
ind[1]
````

````python
ind[::2]
````

`Index`对象还具有许多来自NumPy数组的熟悉属性：

````python
print(ind.size, ind.shape, ind.ndim, ind.dtype)
````

`Index`对象和NumPy数组之间的一个区别是，索引是不可变的——也就是说，它们不能通过正常方式进行修改。

````python
ind[1] = 0
````

这种不可变性使得在多个 ``DataFrame`` 和数组之间共享索引更加安全，避免了因意外修改索引而可能导致的副作用。

### 索引作为有序集合

Pandas 对象旨在简化跨数据集的操作，例如连接，这些操作依赖于集合运算的多个方面。

`Index` 对象遵循 Python 内置 `set` 数据结构使用的许多约定，因此可以以熟悉的方式计算并集、交集、差集和其他组合。

````python
indA = pd.Index([1, 3, 5, 7, 9])
indB = pd.Index([2, 3, 5, 7, 11])
````

````python
indA.intersection(indB)
````

````python
indA.union(indB)
````

````python
indA.symmetric_difference(indB)
````
