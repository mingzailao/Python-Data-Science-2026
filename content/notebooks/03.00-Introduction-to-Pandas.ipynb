# 使用Pandas进行数据处理



在[第2部分](02.00-Introduction-to-NumPy.ipynb)中，我们详细探讨了NumPy及其`ndarray`对象，该对象使得在Python中高效存储和操作密集类型数组成为可能。

在这里，我们将基于这些知识，深入研究Pandas库提供的数据结构。

Pandas是一个建立在NumPy之上的新包，它提供了对`DataFrame`的高效实现。

``DataFrame``本质上是带有附加行和列标签的多维数组，通常包含异构类型和/或缺失数据。

除了为标记数据提供方便的存储接口外，Pandas还实现了一些强大的数据操作，这些操作对于数据库框架和电子表格程序的用户来说都很熟悉。

正如我们所看到的，NumPy的`ndarray`数据结构为数值计算任务中常见的干净、组织良好的数据类型提供了基本特征。

虽然它很好地满足了这一目的，但当我们需要更多灵活性（例如，为数据附加标签、处理缺失数据等）时，其局限性变得明显。此外，在尝试不适合逐元素广播（例如分组、透视等）的操作时，每个方面都是分析周围世界中以多种形式存在的不太结构化的数据的重要组成部分。

Pandas，特别是其`Series`和`DataFrame`对象，在NumPyPy数组结构基础上构建，并有效访问这些占据许多数据科学家时间的数据“清洗”任务。

在本书这一部分，我们将重点关注如何有效使用 `Series`, `DataFrame`, 以及相关结构的机制。

我们将在适当的时候使用真实数据集中的示例，但这些示例并不是我们的重点。
