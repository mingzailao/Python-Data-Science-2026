# 错误与调试：

代码开发和数据分析总是需要一些试验和错误，而IPython包含了简化这一过程的工具。

本节将简要介绍控制Python异常报告的一些选项，并探讨调试代码中错误的工具。

## 控制异常：%xmode

通常情况下，当 Python 脚本失败时，会引发一个异常。

当解释器遇到这些异常之一时，可以在 *traceback* 中找到有关错误原因的信息，该信息可以从 Python 内部访问。

通过 `%xmode` 魔法函数，IPython 允许您控制在引发异常时打印的信息量。

请考虑以下代码：

```python
def func1(a, b):
    return a / b

def func2(x):
    a = x
    b = x - 1
    return func1(a, b)
```

```python
func2(1)
```

调用 `func2` 会导致错误，阅读打印的追踪信息可以让我们准确了解发生了什么。

在默认模式下，这个追踪信息包含几行显示每一步上下文的内容，这些步骤最终导致了错误。

使用 `%xmode` 魔法函数（即*异常模式*），我们可以更改打印的信息。

`%xmode` 接受一个参数，即模式，有三种可能性：`Plain`、`Context` 和 `Verbose`。

默认设置为 `Context`，其输出与刚才展示的一样。

而 `Plain` 模式则更加简洁，但提供的信息较少。

```python
%xmode Plain
```

```python
func2(1)
```

The `Verbose` mode adds some extra information, including the arguments to any functions that are called:

```python
%xmode Verbose
```

```python
func2(1)
```

这段额外的信息可以帮助你更好地理解异常被引发的原因。

那么，为什么不一直使用 `Verbose` 模式呢？

随着代码变得复杂，这种追踪信息可能会非常冗长。

根据具体情况，有时 `Plain` 或 `Context` 模式的简洁性更易于处理。

## 调试：当阅读回溯信息不足以解决问题时

标准的 Python 交互式调试工具是 `pdb`，即 Python 调试器。

该调试器允许用户逐行执行代码，以查看可能导致更复杂错误的原因。

其 IPython 增强版为 `ipdb`，即 IPython 调试器。

启动和使用这两种调试器的方法有很多；我们在这里不会详细介绍。

请参考这两个工具的在线文档以获取更多信息。

在 IPython 中，也许最方便的调试接口是 `%debug` 魔法命令。

如果在遇到异常后调用它，它将自动在异常发生的位置打开一个交互式调试提示符。

`ipdb` 提示符让您可以探索当前堆栈状态、可用变量，并甚至运行 Python 命令！

让我们查看最近的异常，然后进行一些基本操作。我们将打印 `a` 和 `b` 的值，然后输入 `quit` 退出调试会话。

```python
%debug
```

交互式调试器的功能远不止于此——我们甚至可以在调用栈中上下跳转，并探索那里的变量值。


```python
%debug
```

这使我们能够快速找出不仅是导致错误的原因，还有哪些函数调用导致了这个错误。

如果您希望调试器在每次引发异常时自动启动，可以使用 `%pdb` 魔法函数来开启这种自动行为。

```python
%xmode Plain
%pdb on
func2(1)
```

最后，如果您有一个脚本希望从头开始以交互模式运行，可以使用命令 `%run -d`，并使用 `next` 命令逐行交互式地执行代码。

### 调试命令的部分列表

可用于交互式调试的命令远比我在这里展示的要多。以下表格包含了一些更常见和有用命令的描述。

| Command       |  Description                                                |
|---------------|-------------------------------------------------------------|
| `l(ist)`      | Show the current location in the file                       |
| `h(elp)`      | Show a list of commands, or find help on a specific command |
| `q(uit)`      | Quit the debugger and the program                           |
| `c(ontinue)`  | Quit the debugger, continue in the program                  |
| `n(ext)`      | Go to the next step of the program                          |
| `<enter>`     | Repeat the previous command                                 |
| `p(rint)`     | Print variables                                             |
| `s(tep)`      | Step into a subroutine                                      |
| `r(eturn)`    | Return out of a subroutine                                  |

有关更多信息，请在调试器中使用 `help` 命令，或查看 `ipdb` 的 [在线文档](https://github.com/gotcha/ipdb)。
