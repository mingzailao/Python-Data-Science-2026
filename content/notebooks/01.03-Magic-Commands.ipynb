# IPython 魔法命令：


前一章展示了IPython如何高效且交互式地使用和探索Python。

在这里，我们将开始讨论IPython在标准Python语法基础上所增加的一些增强功能。

这些功能在IPython中被称为*魔法命令*，并以`%`字符作为前缀。

这些魔法命令旨在简洁地解决标准数据分析中的各种常见问题。

魔法命令分为两种类型：*行魔法*，用单个`%`前缀表示，仅作用于单行输入；以及*单元魔法*，用双重`%%`前缀表示，可作用于多行输入。

我将在此处演示并讨论一些简短的示例，并稍后回到对几个有用的魔法命令的更深入讨论。

## 运行外部代码：%run

当你开始开发更复杂的代码时，你可能会发现自己在IPython中进行交互式探索，同时使用文本编辑器来存储希望重用的代码。

与其在新窗口中运行这些代码，不如直接在你的IPython会话中运行它，这样更加方便。

这可以通过`%run`魔法命令来实现。

例如，假设你创建了一个名为*myscript.py*的文件，其内容如下：

```python
# file: myscript.py

def square(x):
    """square a number"""
    return x ** 2

for N in range(1, 4):
    print(f"{N} squared is {square(N)}")
```

您可以在您的IPython会话中按如下方式执行此操作：

```ipython
In [6]: %run myscript.py
1 squared is 1
2 squared is 4
3 squared is 9
```

请注意，在运行此脚本后，脚本中定义的任何函数都可以在您的IPython会话中使用。

```ipython
In [7]: square(5)
Out[7]: 25
```

有几种选项可以微调代码的运行方式；您可以通过在IPython解释器中输入**`%run?`**以正常方式查看文档。

## 代码执行时间：%timeit

另一个有用的魔法函数是 `%timeit`，它将自动确定后续单行 Python 语句的执行时间。

例如，我们可能想要检查列表推导式的性能。

```ipython
In [8]: %timeit L = [n ** 2 for n in range(1000)]
430 µs ± 3.21 µs per loop (mean ± std. dev. of 7 runs, 1000 loops each)
```

`%timeit` 的好处在于，对于短命令，它会自动执行多次运行，以获得更可靠的结果。

对于多行语句，添加第二个 `%` 符号将使其变为一个单元魔法，可以处理多行输入。

例如，这里是使用 `for` 循环的等效结构：

```ipython
In [9]: %%timeit
   ...: L = []
   ...: for n in range(1000):
   ...:     L.append(n ** 2)
   ...: 
484 µs ± 5.67 µs per loop (mean ± std. dev. of 7 runs, 1000 loops each)
```

我们可以立即看到，在这种情况下，列表推导式比等效的 `for` 循环结构快约 10%。

## 魔法函数帮助：？，%magic 和 %lsmagic

与普通的 Python 函数一样，IPython 的魔法函数也具有文档字符串，这些有用的文档可以通过标准方式访问。

例如，要查看 `%timeit` 魔法函数的文档，只需输入以下内容：

```ipython
In [10]: %timeit?
```

其他函数的文档可以以类似方式访问。

要获取可用魔法函数的一般描述，包括一些示例，您可以输入以下内容：

```ipython
In [11]: %magic
```

要快速简单地列出所有可用的魔法函数，请输入以下内容：

```ipython
In [12]: %lsmagic
```
