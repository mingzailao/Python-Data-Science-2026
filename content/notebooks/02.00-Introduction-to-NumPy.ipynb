# NumPy简介

本书的这一部分与[第3部分](03.00-Introduction-to-Pandas.ipynb)一起，概述了在Python中有效加载、存储和操作内存数据的技术。

该主题非常广泛：数据集可以来自各种来源，并以多种格式存在，包括文档集合、图像集合、音频片段集合、数值测量集合或几乎任何其他形式。

尽管表面上看似异质，但许多数据集从根本上可以表示为数字数组。

例如，图像——尤其是数字图像——可以简单地视为表示区域内像素亮度的二维数字数组。

音频片段可以被视为强度与时间的一维数组。

文本可以通过多种方式转换为数值表示，例如用二进制数字表示某些单词或单词对的出现频率。

无论数据是什么，使其可分析的第一步都是将其转换为数字数组。

（我们将在[特征工程](05.04-Feature-Engineering.ipynb)中讨论此过程的一些具体示例。）

因此，高效存储和操作数值数组对于进行数据科学至关重要。

接下来，我们将看看Python处理这些数值数组所需的专门工具：NumPy包和Pandas包（在[第3部分](03.00-Introduction-to-Pandas.ipynb)中讨论）。

本书这一部分将详细详细介绍NumPy。NumPy（*Numerical Python* 的缩写）提供了一个高效接口，用于存储和操作密集的数据缓冲区。

在某些方面，NumPy数组类似于Python内置的`list`类型，但随着数组规模增大，NumPy数组提供了更高效的存储和数据操作能力。

NumPy数组构成了几乎整个Python数据科学工具生态系统的核心，因此学习如何有效使用NumPy所花费的时间，无论您对数据科学哪个方面感兴趣，都将是有价值的。
