# NumPy数组的基础知识


在Python中，数据操作几乎与NumPy数组操作同义：即使是像Pandas这样的新工具（[第3部分](03.00-Introduction-to-Pandas.ipynb)）也建立在NumPy数组之上。

本章将展示几个使用NumPy数组操作的示例，以访问数据和子数组，并对数组进行分割、重塑和连接。

虽然这里展示的操作类型可能显得有些枯燥和学究，但它们构成了书中许多其他示例的基础。

我们将在这里涵盖一些基本数组操作的类别：

- *数组属性*：确定数组的大小、形状、内存消耗和数据类型

- *数组索引*：获取和设置单个数组元素的值

- *数组切片*：在较大的数组中获取和设置较小的子阵列

- *重塑阵列*：改变给定阵列的形状

- *连接与拆分阵列*：将多个阵列合并为一个，以及将一个阵列拆分为多个。

## NumPy数组属性：


首先，让我们讨论一些有用的数组属性。

我们将从定义一维、二维和三维的随机数组开始。

我们将使用NumPy的随机数生成器，并用一个固定值对其进行*初始化*，以确保每次运行此代码时生成相同的随机数组。

````python
import numpy as np
rng = np.random.default_rng(seed=1701)  # seed for reproducibility

x1 = rng.integers(10, size=6)  # one-dimensional array
x2 = rng.integers(10, size=(3, 4))  # two-dimensional array
x3 = rng.integers(10, size=(3, 4, 5))  # three-dimensional array
````

每个数组都有一些属性，包括 `ndim`（维度数量）、`shape`（每个维度的大小）、`size`（数组的总大小）和 `dtype`（每个元素的类型）：

````python
print("x3 ndim: ", x3.ndim)
print("x3 shape:", x3.shape)
print("x3 size: ", x3.size)
print("dtype:   ", x3.dtype)
````

更多关于数据类型的讨论，请参见[理解Python中的数据类型](02.01-Understanding-Data-Types.ipynb)。

## 数组索引：访问单个元素


如果您熟悉Python的标准列表索引，那么NumPy中的索引将会显得非常熟悉。

在一维数组中，可以通过在方括号中指定所需的索引来访问第$i^{th}$个值（从零开始计数），这与Python列表的用法相同。

````python
x1
````

````python
x1[0]
````

````python
x1[4]
````

要从数组的末尾进行索引，可以使用负数索引：

````python
x1[-1]
````

````python
x1[-2]
````

在多维数组中，可以使用以逗号分隔的 `(行, 列)` 元组来访问项：

````python
x2
````

````python
x2[0, 0]
````

````python
x2[2, 0]
````

````python
x2[2, -1]
````

值也可以使用任何前面的索引表示法进行修改：

````python
x2[0, 0] = 12
x2
````

请注意，与Python列表不同，NumPy数组具有固定类型。

这意味着，例如，如果您尝试将浮点值插入整数数组，该值将被静默截断。请不要对这种行为感到意外！

````python
x1[0] = 3.14159  # this will be truncated!
x1
````

## 数组切片：访问子数组


正如我们可以使用方括号访问单个数组元素，我们也可以使用带有冒号（`:`）字符的*切片*表示法来访问子数组。

NumPy 的切片语法遵循标准 Python 列表的规则；要访问数组 `x` 的一个切片，可以使用以下格式：

``` python
x[start:stop:step]
```

如果这些参数中的任何一个未指定，则默认为 `start=0`，`stop=<维度大小>`，`step=1`。

接下来，让我们看一些在一维和多维中访问子数组的示例。

### 一维子数组

以下是访问一维子数组中元素的一些示例：

````python
x1
````

````python
x1[:3]  # first three elements
````

````python
x1[3:]  # elements after index 3
````

````python
x1[1:4]  # middle subarray
````

````python
x1[::2]  # every second element
````

````python
x1[1::2]  # every second element, starting at index 1
````

一个可能令人困惑的情况是，当 `step` 值为负时。

在这种情况下，`start` 和 `stop` 的默认值会被交换。

这成为反转数组的一种方便方法：

````python
x1[::-1]  # all elements, reversed
````

````python
x1[4::-2]  # every second element from index 4, reversed
````

### 多维子数组

多维切片的工作方式相同，多个切片之间用逗号分隔。

例如：

````python
x2
````

````python
x2[:2, :3]  # first two rows & three columns
````

````python
x2[:3, ::2]  # three rows, every second column
````

````python
x2[::-1, ::-1]  # all rows & columns, reversed
````

#### 访问数组的行和列

一种常见的操作是访问数组的单个行或列。

这可以通过结合索引和切片来实现，使用一个由单个冒号（`:`）标记的空切片。

````python
x2[:, 0]  # first column of x2
````

````python
x2[0, :]  # first row of x2
````

在行访问的情况下，可以省略空切片，以实现更紧凑的语法：

````python
x2[0]  # equivalent to x2[0, :]
````

### 子数组作为无拷贝视图

与 Python 列表切片不同，NumPy 数组切片返回的是数组数据的 *视图* 而不是 *拷贝*。

请考虑我们之前提到的二维数组：

````python
print(x2)
````

让我们从中提取一个 $2 \times 2$ 子数组： 


````python
x2_sub = x2[:2, :2]
print(x2_sub)
````

现在，如果我们修改这个子数组，我们会发现原始数组发生了变化！请注意：

````python
x2_sub[0, 0] = 99
print(x2_sub)
````

````python
print(x2)
````

一些用户可能会对此感到惊讶，但这实际上是有利的：例如，在处理大型数据集时，我们可以访问并处理这些数据集中的部分内容，而无需复制底层数据缓冲区。

### 创建数组的副本

尽管数组视图具有良好的特性，但有时显式地复制数组或子数组中的数据是很有用的。这可以通过 `copy` 方法最简单地实现。

````python
x2_sub_copy = x2[:2, :2].copy()
print(x2_sub_copy)
````

如果我们现在修改这个子数组，原始数组不会受到影响：

````python
x2_sub_copy[0, 0] = 42
print(x2_sub_copy)
````

````python
print(x2)
````

## 数组的重塑

另一种有用的操作是数组的重塑，这可以通过 `reshape` 方法来实现。

例如，如果您想将数字 1 到 9 放入一个 $3 \times 3$ 的网格中，可以执行以下操作：

````python
grid = np.arange(1, 10).reshape(3, 3)
print(grid)
````

Note that for this to work, the size of the initial array must match the size of the reshaped array, and in most cases the `reshape` method will return a no-copy view of the initial array.

常见的重塑操作是将一维数组转换为二维行或列矩阵。

````python
x = np.array([1, 2, 3])
x.reshape((1, 3))  # row vector via reshape
````

````python
x.reshape((3, 1))  # column vector via reshape
````

方便的简写方式是使用 `np.newaxis` 在切片语法中：

````python
x[np.newaxis, :]  # row vector via newaxis
````

````python
x[:, np.newaxis]  # column vector via newaxis
````

这是我们在本书剩余部分中将经常使用的一种模式。

## 数组连接与拆分

前述所有例程均针对单个数组进行操作。NumPy 还提供了将多个数组合并为一个的工具，并可以将单个数组拆分为多个数组。

### 数组的连接

在NumPy中，数组的连接或合并主要通过`np.concatenate`、`np.vstack`和`np.hstack`这几种方法来实现。

`np.concatenate`将一个元组或数组列表作为其第一个参数，如下所示：

````python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
````

您还可以一次连接多个数组：

````python
z = np.array([99, 99, 99])
print(np.concatenate([x, y, z]))
````

它可以用于二维数组：

````python
grid = np.array([[1, 2, 3],
                 [4, 5, 6]])
````

````python
# concatenate along the first axis
np.concatenate([grid, grid])
````

````python
# concatenate along the second axis (zero-indexed)
np.concatenate([grid, grid], axis=1)
````

对于处理混合维度的数组，使用 `np.vstack`（垂直堆叠）和 `np.hstack`（水平堆叠）函数会更加清晰。

````python
# vertically stack the arrays
np.vstack([x, grid])
````

````python
# horizontally stack the arrays
y = np.array([[99],
              [99]])
np.hstack([grid, y])
````

同样，对于高维数组，`np.dstack` 将沿着第三个轴堆叠数组。

### 数组的拆分

连接的反义词是拆分，这可以通过函数 `np.split`、`np.hsplit` 和 `np.vsplit` 来实现。对于每个函数，我们可以传递一个包含拆分点的索引列表。

````python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3 = np.split(x, [3, 5])
print(x1, x2, x3)
````

注意，*N* 个分割点会导致 *N* + 1 个子数组。

相关的函数 `np.hsplit` 和 `np.vsplit` 是相似的：

````python
grid = np.arange(16).reshape((4, 4))
grid
````

````python
upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)
````

````python
left, right = np.hsplit(grid, [2])
print(left)
print(right)
````

同样，对于高维数组，`np.dsplit` 将沿着第三个轴分割数组。
