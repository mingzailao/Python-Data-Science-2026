# Getting Started in IPython and Jupyter

在编写用于数据科学的Python代码时，我通常会在三种工作模式之间切换：我使用IPython shell来尝试短序列命令，使用Jupyter Notebook进行更长的交互式分析和与他人共享内容，以及使用像Emacs或VSCode这样的交互式开发环境（IDE）来创建可重用的Python包。

本章重点讨论前两种模式：IPython shell和Jupyter Notebook。

虽然IDE在软件开发中是数据科学家工具箱中的重要第三个工具，但我们将在此不直接讨论。

## 启动 IPython Shell

## 启动 Jupyter Notebook

Jupyter Notebook 是一个基于浏览器的图形界面，用于 IPython shell，并在其基础上构建了一套丰富的动态显示功能。

除了执行 Python/IPython 语句外，Notebook 还允许用户包含格式化文本、静态和动态可视化、数学方程式、JavaScript 小部件等多种内容。

此外，这些文档可以以一种方式保存，使其他人能够打开并在自己的系统上执行代码。

尽管您将通过网页浏览器窗口查看和编辑 Jupyter Notebook，但它们必须连接到正在运行的 Python 进程才能执行代码。

您可以通过在系统终端中运行以下命令来启动此过程（称为“内核”）：

```
$ jupyter lab
```

该命令将启动一个本地 Web 服务器，您的浏览器将能看到它。

它会立即输出一条日志，显示正在进行的操作；该日志大致如下所示：

```
$ jupyter lab

[ServerApp] 从本地目录提供笔记本：/Users/jakevdp/PythonDataScienceHandbook

[ServerApp] Jupyter Server 1.4.1 正在运行于：

[ServerApp] http://localhost:8888/lab?token=dd852649

[ServerApp] 使用 Control-C 停止此服务器并关闭所有内核（按两次以跳过确认）。
```

`

发出命令后，您的默认浏览器应自动打开并导航至列出的本地 URL；具体地址取决于您的系统。

如果浏览器未自动打开，您可以手动打开一个窗口并输入此地址（例如 *http://localhost:8888/lab/*）。

## Help and Documentation in IPython

如果你在本章中只阅读一个部分，请阅读这一部分：我认为这里讨论的工具是IPython对我日常工作流程最具变革性的贡献。

在数据科学领域也是如此：可搜索的网络资源，如在线文档、邮件列表线程和Stack Overflow上的回答，包含了丰富的信息，即使（尤其？）是关于你之前曾经搜索过的话题。

成为一名有效的数据科学从业者，不仅仅是记住每种可能情况应使用的工具或命令，更重要的是学会有效地寻找自己不知道的信息，无论是通过网络搜索引擎还是其他方式。

IPython/Jupyter最有用的功能之一就是缩短用户与能够帮助他们高效工作的文档和搜索之间的距离。

虽然网络搜索仍然在解答复杂问题中发挥作用，但通过IPython单独可以找到大量信息。

以下是一些IPython可以通过几次击键来帮助回答的问题示例：

- 我该如何调用这个函数？它有哪些参数和选项？

- 这个Python对象的源代码是什么样子的？

- 我导入的这个包里有什么？

- 这个对象具有哪些属性或方法？

接下来，我们将讨论IPython shell和Jupyter Notebook提供的一些工具，以便快速访问访问这些信息，包括`?`字符用于探索文档，`??`字符用于查看源代码，以及Tab键用于自动补全。

### 通过 ? 访问文档

Python语言及其数据科学生态系统以用户为中心，其中一个重要部分是对文档的访问。

每个Python对象都包含一个字符串引用，称为*docstring*，在大多数情况下，它将包含该对象的简要总结以及如何使用它的信息。

Python提供了一个内置的`help`函数，可以访问这些信息并打印结果。

例如，要查看内置`len`函数的文档，可以执行以下操作：

```ipython
In [1]: help(len)

Help on built-in function len in module builtins:

len(obj, /)

Return the number of items in a container.
```

根据您的解释器，这些信息可能会作为行内文本显示，也可能在单独的弹出窗口中显示。

因为在对象上寻找帮助是如此普遍且有用，IPython 和 Jupyter 引入了 `?` 字符作为访问此文档和其他相关信息的简写方式。

```ipython
In [2]: len?
Signature: len(obj, /)
Docstring: Return the number of items in a container.
Type:      builtin_function_or_method
```

该符号几乎适用于任何事物，包括对象方法：

```ipython
In [3]: L = [1, 2, 3]
In [4]: L.insert?
Signature: L.insert(index, object, /)
Docstring: Insert object before index.
Type:      builtin_function_or_method
```

或甚至是对象本身，附带其类型的文档：

```ipython
In [5]: L?
Type:        list
String form: [1, 2, 3]
Length:      3
Docstring:  
Built-in mutable sequence.

If no argument is given, the constructor creates a new empty list.
The argument must be an iterable if specified.
```

重要的是，这甚至适用于您自己创建的函数或其他对象！

在这里，我们将定义一个带有文档字符串的小函数：

```ipython
In [6]: def square(a):
  ....:     """Return the square of a."""
  ....:     return a ** 2
  ....:
```

请注意，为了为我们的函数创建文档字符串，我们只需在第一行放置一个字符串字面量。

因为文档字符串通常是多行的，按照惯例，我们使用了Python的三重引号表示多行字符串。

现在我们将使用 `?` 来查找这个文档字符串：

```ipython
In [7]: square?
Signature: square(a)
Docstring: Return the square of a.
File:      <ipython-input-6>
Type:      function
```

这种通过文档字符串快速访问文档的方式，是你应该养成在编写代码时始终添加此类内联文档习惯的一个原因！

### 通过双问号访问源代码

由于Python语言的可读性很高，通常可以通过阅读您感兴趣对象的源代码获得更深层次的见解。

IPython和Jupyter提供了一个快捷方式，通过双问号（`??`）来访问源代码。

```ipython
In [8]: square??
Signature: square(a)
Source:   
def square(a):
    """Return the square of a."""
    return a ** 2
File:      <ipython-input-6>
Type:      function
```
对于像这样的简单函数，双问号可以快速洞察其内部细节。

如果你多加尝试，你会注意到有时 `??` 后缀不会显示任何源代码：这通常是因为相关对象并不是用 Python 实现的，而是用 C 或其他编译扩展语言实现的。

如果是这种情况，`??` 后缀的输出与 `?` 后缀相同。

你会发现这一点在许多 Python 的内置对象和类型中尤为明显，包括之前提到的 `len` 函数。

```ipython
In [9]: len??
Signature: len(obj, /)
Docstring: Return the number of items in a container.
Type:      builtin_function_or_method
```

使用 `?` 和/或 `??` 是快速有效地获取有关任何 Python 函数或模块功能信息的强大方式。

### 探索带有 Tab 补全的模块

另一个有用的接口是使用 Tab 键进行自动补全和探索对象、模块及命名空间的内容。

在接下来的示例中，我将使用 `<TAB>` 来指示何时应按下 Tab 键。

#### 对象内容的Tab补全

每个Python对象都有各种属性和方法与之相关。

如前所述的`help`函数，Python还内置了一个`dir`函数，它返回这些属性和方法的列表，但在实际使用中，Tab补全界面要容易得多。

要查看对象所有可用属性的列表，可以输入对象名称后跟一个句点（“`.`”）字符，然后按下Tab键。

```ipython
In [10]: L.<TAB>
            append() count    insert   reverse 
            clear    extend   pop      sort    
            copy     index    remove           
```

为了缩小列表，您可以输入名称的第一个字符或多个字符，然后按 Tab 键将找到匹配的属性和方法。

```ipython
In [10]: L.c<TAB>
             clear() count()
             copy()         

In [10]: L.co<TAB>
              copy()  count()
```

如果只有一个选项，按下 Tab 键将为您完成该行。

例如，以下内容将立即替换为 `L.count`：

```ipython
In [10]: L.cou<TAB>

```

尽管Python没有严格区分公共/外部属性和私有/内部属性，但根据约定，前面加下划线用于表示后者。

为了清晰起见，这些私有方法和特殊方法默认情况下被省略在列表中，但可以通过显式输入下划线来列出它们。

```ipython
In [10]: L._<TAB>
           __add__             __delattr__     __eq__      
           __class__           __delitem__     __format__()
           __class_getitem__() __dir__()       __ge__            >
           __contains__        __doc__         __getattribute__     
```

为了简洁起见，我只展示了输出的前几列。

其中大多数是Python的特殊双下划线方法（通常被称为“dunder”方法）。

#### 导入时的Tab补全

在从包中导入对象时，Tab补全也非常有用。

在这里，我们将使用它来查找`itertools`包中以`co`开头的所有可能导入。

```ipython
In [10]: from itertools import co<TAB>
         combinations()                  compress()
         combinations_with_replacement() count()
```

同样，您可以使用标签补全功能查看系统中可用的导入项（这将根据哪些第三方脚本和模块在您的Python会话中可见而有所不同）。

```ipython
In [10]: import <TAB>
            abc                 anyio                          
            activate_this       appdirs                        
            aifc                appnope        >
            antigravity         argon2                         

In [10]: import h<TAB>
            hashlib html   
            heapq   http   
            hmac        
```

#### 超越标签补全：通配符匹配

标签补全在你知道所需对象或属性名称的前几个字符时非常有用，但如果想要匹配名称中间或末尾的字符，则帮助不大。

对于这种使用情况，IPython 和 Jupyter 提供了一种使用 `*` 字符进行名称通配符匹配的方法。

例如，我们可以利用这一点列出命名空间中所有以 `Warning` 结尾的对象。

```ipython
In [10]: *Warning?
BytesWarning                  RuntimeWarning
DeprecationWarning            SyntaxWarning
FutureWarning                 UnicodeWarning
ImportWarning                 UserWarning
PendingDeprecationWarning     Warning
ResourceWarning
```
注意，`*` 字符可以匹配任何字符串，包括空字符串。

同样，假设我们正在寻找一个方法名中包含单词 `find` 的字符串方法。

我们可以这样进行搜索：

```ipython
In [11]: str.*find*?
str.find
str.rfind
```

我发现这种灵活的通配符搜索在了解新包或重新熟悉一个熟悉的包时，对于查找特定命令非常有用。
