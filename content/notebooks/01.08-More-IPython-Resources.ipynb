# 更多 IPython 资源：


在这一系列章节中，我们仅仅触及了使用IPython进行数据科学任务的表面。

无论是在印刷版还是网络上，都有更多的信息可供获取。在这里，我将列出一些您可能会觉得有用的其他资源。

## 网络资源

- [IPython官方网站](http://ipython.org)：IPython官方网站提供文档、示例、教程以及各种其他资源的链接。

- [nbviewer网站](http://nbviewer.jupyter.org/)：该网站展示了互联网上任何可用Jupyter笔记本的静态渲染。首页上有一些示例笔记本，您可以浏览以了解其他人如何使用IPython！

- [精心策划的Jupyter笔记本集合](https://github.com/jupyter/jupyter/wiki)：这个不断增长的笔记本列表由nbviewer支持，展示了您可以使用IPython进行数值分析的深度和广度。它包括从简短示例和教程到完整课程和以笔记本格式编写的书籍的一切内容！

- 视频教程：在互联网上搜索，您会发现许多关于IPython的视频教程。我特别推荐查找Fernando Perez和Brian Granger在PyCon、SciPy和PyData会议上的讲座，他们是IPython和Jupyter的主要创始人及维护者。

## 书籍

- [*Python for Data Analysis* (O'Reilly)](http://shop.oreilly.com/product/0636920023784.do)：Wes McKinney的书中包含了一章，介绍了如何将IPython作为数据科学家使用。尽管许多内容与我们在这里讨论的重叠，但另一个视角总是有帮助的。

- [*Learning IPython for Interactive Computing and Data Visualization* (Packt)](https://www.packtpub.com/big-data-and-business-intelligence/learning-ipython-interactive-computing-and-data-visualization)：Cyrille Rossant撰写的这本短小精悍的书为使用IPython进行数据分析提供了良好的入门。

- [*IPython Interactive Computing and Visualization Cookbook* (Packt)](https://www.packtpub.com/big-data-and-business-intelligence/ipython-interactive-computing-and-visualization-cookbook)：同样由Cyrille Rossant编写，这本书对使用IPython进行数据科学进行了更长且更深入的探讨。尽管其名称如此，它不仅仅涉及IPython，还深入探讨了广泛的数据科学主题。

最后，提醒您可以自行寻找帮助：如果您善用并经常使用，基于`?`的帮助功能（详见[Help and Documentation in IPython](01.01-Help-And-Documentation.ipynb)）会非常有用。

。

在您浏览此处及其他地方的示例时，可以利用这些资源熟悉IPython所提供的所有工具。
