# IPython 和 Shell 命令：

在与标准Python解释器进行交互时，令人沮丧的一点是需要在多个窗口之间切换，以访问Python工具和系统命令行工具。

IPython弥补了这一缺口，并为您提供了一种直接从IPython终端执行shell命令的语法。

神奇之处在于感叹号：出现在`!`后面的任何内容都将由系统命令行而非Python内核执行。

以下讨论假设您使用的是类Unix系统，例如Linux或macOS。

接下来的某些示例将在Windows上失败，因为Windows默认使用不同类型的shell，但如果您使用*Windows Subsystem for Linux*，这里的示例应该可以正常运行。

如果您对shell命令不熟悉，我建议查看由优秀的软件工程协会（Software Carpentry Foundation）整理的[Unix shell教程](http://swcarpentry.github.io/shell-novice/)。

## 快速介绍 Shell

对使用 shell/终端/命令行的全面介绍超出了本章的范围，但对于初学者，我将在这里提供一个简要的介绍。

Shell 是一种与计算机进行文本交互的方式。

自从1980年代中期，微软和苹果推出了它们现在无处不在的图形操作系统的第一个版本以来，大多数计算机用户通过熟悉的菜单选择和拖放操作与他们的操作系统互动。

然而，在这些图形用户界面出现之前，操作系统早已存在，并主要通过一系列文本输入来控制：在提示符下，用户输入命令，计算机会按照用户指示执行相应操作。

那些早期提示系统是如今大多数数据科学家仍然使用的 shell 和终端的前身。

不熟悉 shell 的人可能会问，当许多相同结果可以通过简单地点击图标和菜单实现时，你为什么还要费心去学习这个？

而 shell 用户可能会反问：既然你可以更轻松地通过键入来完成任务，为何还要寻找图标和菜单项呢？

虽然这听起来像是典型技术偏好之间的一种僵局，但当超越基本任务时，很快就会发现 shell 在高级任务上提供了更多控制——尽管学习曲线确实令人望而生畏。

作为例子，这里有一个 Linux/macOS shell 会话示例例，其中用户探索、创建并修改其系统上的目录和文件（`osx:~ $` 是提示符，而 `$` 后面的内容是输入命令；以 `#` 开头的文本仅作描述，而不是实际需要输入的信息）：

```bash
osx:~ $ echo "hello world"             # echo is like Python's print function
hello world

osx:~ $ pwd                            # pwd = print working directory
/home/jake                             # This is the "path" that we're sitting in

osx:~ $ ls                             # ls = list working directory contents
notebooks  projects 

osx:~ $ cd projects/                   # cd = change directory

osx:projects $ pwd
/home/jake/projects

osx:projects $ ls
datasci_book   mpld3   myproject.txt

osx:projects $ mkdir myproject          # mkdir = make new directory

osx:projects $ cd myproject/

osx:myproject $ mv ../myproject.txt ./  # mv = move file. Here we're moving the
                                        # file myproject.txt from one directory
                                        # up (../) to the current directory (./).
osx:myproject $ ls
myproject.txt
```
请注意，以上内容只是通过输入命令而非点击图标和菜单，以紧凑的方式执行熟悉的操作（如导航目录结构、创建目录、移动文件等）。

仅凭几个命令（`pwd`、`ls`、`cd`、`mkdir` 和 `cp`），您就可以完成许多最常见的文件操作。然而，当您超越这些基础知识时，Shell 方法才真正展现出其强大之处。

## IPython中的Shell命令

任何标准的Shell命令都可以通过在前面加上`!`字符直接在IPython中使用。

例如，可以如下运行`ls`、`pwd`和`echo`命令：

```ipython
In [1]: !ls

myproject.txt

In [2]: !pwd

/home/jake/projects/myproject

In [3]: !echo "从Shell打印"

从Shell打印
```



## 从Shell传递值

Shell命令不仅可以从IPython调用，还可以与IPython命名空间进行交互。

例如，您可以使用赋值运算符`=`将任何Shell命令的输出保存到Python列表中：

```ipython
In [4]: contents = !ls

In [5]: print(contents)

['myproject.txt']

In [6]: directory = !pwd

In [7]: print(directory)

['/Users/jakevdp/notebooks/tmp/myproject']
```

这些结果并不是以列表形式返回，而是作为在IPython中定义的一种特殊Shell返回类型：

```ipython
In [8]: type(directory)

IPython.utils.text.SList
```

这看起来和行为都很像Python列表，但具有额外的功能，例如`grep`和`fields`方法，以及`s`、`n`和`p`属性，这些允许您以方便的方式搜索、过滤和显示结果。

有关更多信息，您可以使用IPython内置的帮助功能。

在另一方向的通信——将Python变量传递到shell中——可以使用`{varname}`语法实现：

```ipython
In [9]: message = "hello from Python"

In [10]: !echo {message}

hello from Python
```

花括号内包含变量名，在shell命令中会被该变量的内容替换。

## 与Shell相关的魔法命令

如果你玩IPython的shell命令一段时间，你可能会注意到无法使用 `!cd` 来导航文件系统：

```ipython
In [11]: !pwd
/home/jake/projects/myproject

In [12]: !cd ..
In [13]: !pwd
/home/jake/projects/myproject
```

原因在于，笔记本中的shell命令是在一个临时子shell中执行的，该子shell不会保持从一个命令到下一个命令的状态。

如果你想以更持久的方式更改工作目录，可以使用 `%cd` 魔法命令：

```ipython
In [14]: %cd ..
/home/jake/projects
```

实际上，默认情况下，你甚至可以不带 `%` 符号直接使用它：

```ipython
In [15]: cd myproject
/home/jake/projects/myproject
```

这被称为 *自动魔法* 功能，而执行这些没有显式 `%` 的命令的能力可以通过 `%automagic` 魔法函数进行切换。

除了 `%cd` 之外，还有其他可用的类似shell的魔法函数，包括 `%cat`, `%cp`, `%env`, `%ls`, `%man`, `%mkdir`, `%more`, `%mv`, `%pwd`, `%rm` 和 ％rmdir，这些都可以在 `automagic` 开启时无需前缀符号使用。

这样一来，你几乎可以可以将IPython提示视为普通shell：

```ipython
In [16]: mkdir tmp

In [17]: ls 
myproject.txt tmp/

In [18]: cp myproject.txt tmp/

In [19]: ls tmp 
myproject.txt 

In [20]: rm -r tmp 
```

这种在与Python会话相同终端窗口内访问shell的方法，使得你能够更加自然地将Python和shell结合在你的工作流程中，从而减少上下文切换。
