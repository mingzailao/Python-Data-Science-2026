# IPython Shell中的键盘快捷键

如果你在电脑上花费了一定时间，可能已经发现了键盘快捷键在工作流程中的用途。

最为人所熟知的可能是 Cmd-c 和 Cmd-v（或 Ctrl-c 和 Ctrl-v），它们用于在各种程序和系统中进行复制和粘贴。

高级用户往往更进一步：流行的文本编辑器如 Emacs、Vim 等，通过复杂的按键组合为用户提供了令人难以置信的操作范围。

IPython shell 没有达到这种程度，但确实提供了一些键盘快捷键，以便在输入命令时快速导航。

虽然这些快捷键中的一些可以在基于浏览器的笔记本中使用，但本节主要讨论 IPython shell 中的快捷键。

一旦你习惯了这些，它们将非常有助于快速执行某些命令，而无需将手从“主”位置移动开来。

如果你是 Emacs 用户，或者对 Linux 风格的 shell 有经验，那么以下内容会让你感到非常熟悉。

我将把这些快捷键分成几个类别：*导航快捷键*、*文本输入快捷键*、*命令历史快捷键*以及 *其他杂项快捷键信息*。

## 导航快捷键

虽然使用左箭头和右箭头键在行中向后和向前移动是显而易见的，但还有其他选项，无需将手从“主”键盘位置移开。

| Keystroke                       | Action                                     |
|---------------------------------|--------------------------------------------|
| Ctrl-a                          | 将光标移动到行首：                            |
| Ctrl-e                          | 将光标移动到行末：                            |
| Ctrl-b or the left arrow key    | 将光标向后移动一个字符。                       |
| Ctrl-f or the right arrow key   | 将光标向前移动一个字符：                       |

## 文本输入快捷键

虽然每个人都熟悉使用退格键删除前一个字符，但伸手去按这个键往往需要一些小的手指灵活性，而且它一次只能删除一个字符。

在IPython中，有几种快捷方式可以删除您正在输入文本的一部分；其中最实用的是用于删除整行文本的命令。

如果您发现自己使用Ctrl-b和Ctrl-d的组合，而不是伸手去按退格键来删除前一个字符，那么您就知道这些操作已经变成了您的第二天性！

| Keystroke                   | Action                                           |
|-----------------------------|--------------------------------------------------|
| Backspace key               | Delete previous character in line                |
| Ctrl-d                      | Delete next character in line                    |
| Ctrl-k                      | Cut text from cursor to end of line              |
| Ctrl-u                      | Cut text from beginning of line to cursor        |
| Ctrl-y                      | Yank (i.e., paste) text that was previously cut  |
| Ctrl-t                      | Transpose (i.e., switch) previous two characters |

## 命令历史快捷键

这里讨论的最具影响力的快捷键可能是 IPython 提供的用于浏览命令历史的功能。

该命令历史不仅限于您当前的 IPython 会话：您的整个命令历史存储在 IPython 配置目录中的 SQLite 数据库中。

访问之前命令的最简单方法是使用上箭头和下箭头键逐步浏览历史，但也存在其他选项。

| Keystroke                         | Action                                     |
|-----------------------------------|--------------------------------------------|
| Ctrl-p (or the up arrow key)      | Access previous command in history         |
| Ctrl-n (or the down arrow key)    | Access next command in history             |
| Ctrl-r                            | Reverse-search through command history     |

反向搜索选项特别有用。

回想一下，我们之前定义了一个名为 `square` 的函数。

现在，让我们从新的 IPython shell 中反向搜索我们的 Python 历史记录，再次找到这个定义。

当你在 IPython 终端中按下 Ctrl-r 时，你会看到以下提示：

```ipython
In [1]:
(reverse-i-search)`': 
```

如果您在此提示符下开始输入字符，IPython 将自动填充与这些字符匹配的最近命令（如果有的话）。

```ipython
In [1]: 
(reverse-i-search)`sqa': square??
```

在任何时候，您都可以添加更多字符以细化搜索，或者再次按 Ctrl-r 以进一步搜索与查询匹配的其他命令。如果您之前跟随操作，再次按两次 Ctrl-r 将会得到：

```ipython
In [1]: 
(reverse-i-search)`sqa': def square(a):
    """Return the square of a"""
    return a ** 2
```

一旦找到所需的命令，请按回车键，搜索将结束。

然后您可以使用检索到的命令继续您的会话。

```ipython
In [1]: def square(a):
    """Return the square of a"""
    return a ** 2

In [2]: square(2)
Out[2]: 4
```

请注意，您可以使用 Ctrl-p/Ctrl-n 或上下箭头键以类似的方式搜索历史记录，但仅限于匹配行首字符。

也就是说，如果您输入 **`def`** 然后按 Ctrl-p，它将找到您历史记录中以字符 `def` 开头的最新命令（如果有的话）。

## 杂项快捷键

最后，还有一些杂项快捷键，它们不属于前面提到的任何类别，但仍然值得了解。

| Keystroke                   | Action                                     |
|-----------------------------|--------------------------------------------|
| Ctrl-l                      | Clear terminal screen                      |
| Ctrl-c                      | Interrupt current Python command           |
| Ctrl-d                      | Exit IPython session                       |

“特别是，当您不小心启动一个非常长时间运行的任务时，Ctrl-c快捷键可以非常有用。”

虽然这里讨论的一些快捷键起初可能显得有些晦涩，但经过练习，它们很快就会变得自动化。

一旦你培养了这种肌肉记忆，我怀疑你甚至会希望它们在其他场合也能使用。
